```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_to_file = os.path.join("Complete.csv")
deaths_vs_pop = pd.read_csv(path_to_file)
deaths_vs_pop.head(52)
```

```python
state_group = deaths_vs_pop.groupby('State')
state_pop = state_group["Pop2018"].mean()
state_pop.head(51)
```

```python
state_group = deaths_vs_pop.groupby('State')
state_deaths = state_group['Number of Deaths'].mean()
state_deaths.head(51)
```

```python
deaths_vs_pop.reset_index().plot(
    x="State", y="Number of Deaths", kind="bar", color='#8B0000')

plt.title("COVID-19 Deaths Per State")
plt.xlabel("State")
plt.ylabel("Covid-19 Deaths")

plt.savefig("Covid19 Death per state")
```

```python
deaths_vs_pop.reset_index().plot(
    x="State", y="Pop2018", kind="bar", color='r')

plt.title("Population Per State (2018 Pre-Covid19)")
plt.xlabel("State")
plt.ylabel("Population")

plt.savefig("PopulationPerState_PreCovid19")
```

```python
import matplotlib.pyplot as plt
path_to_file = os.path.join("Covid19_states.csv")
deaths_vs_pop_dens = pd.read_csv(path_to_file)
deaths_vs_pop_dens.head(52)
```

```python
state_group = deaths_vs_pop_dens.groupby('State')
state_popdens = state_group[""].mean()
state_popdens.head(51)
```

```python
state_group = deaths_vs_pop_dens.groupby('State')
state_deaths = state_group['Deaths'].mean()
state_deaths.head(51)
```

```python
plt.figure(figsize=(15,4))
x_axis = np.arange(len(deaths_vs_pop_dens))
x_axis
plt.plot(x_axis, deaths_vs_pop_dens["Deaths"], color='#8B0000')
x = plt.xticks(x_axis, deaths_vs_pop_dens["State"], rotation="vertical")
plt.title("Covid-19 Deaths per State (2020)")
plt.xlabel("State")
plt.ylabel("Number of Deaths")

plt.show()
```

```python
plt.figure(figsize=(15,4))
x_axis = np.arange(len(deaths_vs_pop_dens))
x_axis
plt.plot(x_axis, deaths_vs_pop_dens["Pop Density"], color='r')
x = plt.xticks(x_axis, deaths_vs_pop_dens["State"], rotation="vertical")
plt.title("Population Density (Per Capita)")
plt.xlabel("State")
plt.ylabel("Pop. Density")

plt.show()
```
